```python
%matplotlib inline

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

### Loading the data

```python
df = pd.read_csv("datasets/4 Matplotlib/data/table_i702t60.csv")
df.head()
```

```python
# Checking for the missing values in the dataset
df.info()
```

```python
month_number = df["month"].values
interest_paid = df["interest_paid"].values
principal_paid = df["principal_paid"].values
```

```python
plt.figure(figsize=(10,5))

plt.plot(interest_paid,interest_paid, c="k", label="Interest")
plt.plot(interest_paid, principal_paid, c="b", label="Principal")
plt.xlim(left=1, right=225)
plt.ylim(bottom=0, top=1000)
plt.xlabel("Month", fontsize=22)
plt.ylabel("Dollars",fontsize=22)
plt.title("Interest and Principal paid each month",fontsize=26)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(loc=(1.02,0), borderaxespad=0, fontsize=20)
plt.tight_layout()

plt.savefig("graphs/first_plot.png",  dpi=300)
```
